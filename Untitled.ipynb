```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# To combine 2 files or datasets we use pd.merge()
final_df=pd.merge(df,df_country,on='Country Code')
final_df.head(2)
```

```python
## To check data types
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

From this we can say that zomato has highest outlet in india than any other country.

```python
country_names  = final_df.Country.value_counts().index
country_values = final_df.Country.value_counts().values
```

```python
## Pie chart - Top 3 countries that use zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
```

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

Observation
1. Whenever rating is between 4.5 to 4.9 ---> Excellent
2. When ratings are between 4.0 to 4.4 ---> Very good
3. If rating is btwn 3.5 to 3.9 ---> Good
4. if rating is btwn 2.5 to 3.4 ---> Average
5. If rating<2.5 ---> Poor

```python
ratings.head()
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",data = ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data = ratings,palette = ['White','red','orange','yellow','green','green'])
```

Observation:
1. Not rated count is high
2. Max ratings are between 2.5 to 3.4

```python
## count plot
sns.countplot(x="Rating color",data=ratings,hue='Rating color',palette = ['White','red','orange','yellow','green','green'])
```

```python
### Find the countries that has given 0 rating
final_df[final_df['Rating color']=='White']['Country'].value_counts().reset_index()
```

Observations:
1. Max. number of zero ratings are from Indian customers

```python
## find out which currency is used by which country?
final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```

```python
## Which countries do have online delivery option
final_df[final_df['Has Online delivery']=='Yes'].Country.value_counts().reset_index()
```

observations:
1. Online deliveries are avalible in India and UAE
